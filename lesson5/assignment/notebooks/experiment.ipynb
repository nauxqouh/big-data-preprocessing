```python
def divide_age_group(df: pd.DataFrame) -> pd.DataFrame:
    df['age_group'] = pd.cut(df['age'],
                             bins=[28, 37, 45, 53, 61, 69, 77],
                             labels=['29-37', '38-45', '46-53', '54-61', '62-69', '70-77'])
    return df


eda = divide_age_group(df.copy())
eda.head(10)
```

```python
age_group = eda.groupby(['age_group', 'cp']).size().reset_index(name='counts')
age_group.cp = age_group.cp.map({
        0: 'Typical',
        1: 'Atypical',
        2: 'Non-anginal',
        3: 'Asymptomatic'
    })
age_group
```

```python
pivot_table = pd.pivot_table(age_group,
                             values='counts',
                             index='cp',
                             columns='age_group').rank(ascending=False, axis=0)
pivot_table
```
